```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = plt.imread("./5.jpg")
img = img.astype('float32') / 255.
mask = plt.imread("./5_gt.png")
plt.imshow(img)
```

```python
mask.shape
```

```python
n, m, _ = mask.shape
indexTamp = []
for i in range(n):
    for j in range(m):
        if mask[i, j, 0] == 1:
            indexTamp.append((i, j))
```

```python
for sig in range(0, 30, 1):
    noisy = np.copy(img)
    mu, sigma = 0, sig/100
    for i, j in indexTamp:
        noise = np.random.normal(mu, sigma, 3)
        noisy[i, j,:] += noise
        for ch in range(3):
            if noisy[i, j, ch] < 0:
                noisy[i, j, ch] = 0
            elif noisy[i, j, ch] > 1:
                noisy[i, j, ch] = 1
    plt.imsave("./5_{}.jpg".format(sigma), noisy, format='jpg')
```
