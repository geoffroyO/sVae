```python
tabs = np.abs(features-reconstruction)
n, m, _ = tabs.shape
list_comp = []
for i in range(n):
    for j in range(m):
        elt = tabs[i, j, :]
        list_comp.append(elt)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
tabs = pca.fit_transform(list_comp)
```

```python
print(pca.explained_variance_ratio_)
```
